> Data Cleaning

```python
#import library for proprocessing
#and Intanciating

import nltk
stopwords=nltk.corpus.stopwords.words('english')
lemmitizer=nltk.stem.WordNetLemmatizer()
import string
```

```python
#Data Cleaning using stopswords,string and Lemmetizer


eng_df['original_title']=eng_df['original_title'].apply(lambda X: X.translate(str.maketrans(' ',' ',string.punctuation)))
eng_df['original_title']=eng_df['original_title'].apply(lambda X: ' '.join([i for i in X.split()if i not in stopwords]))
eng_df['original_title']=eng_df['original_title'].apply(lambda X: ' '.join([lemmitizer.lemmatize(i)for i in X.split()]))
```

```python
#converting it to small capital

eng_df['original_title']=eng_df['original_title'].str.lower()
```

```python
#then converting it to number


from sklearn.feature_extraction.text import CountVectorizer 

cv=CountVectorizer()
cv.fit(eng_df['original_title']) 
cv_model=cv.transform(eng_df['original_title'])
```

```python
#when our ojective is topic modeeling,the for LatentDirichlet for modelling


from sklearn.decomposition import LatentDirichletAllocation
```

```python
#initiializing

model=LatentDirichletAllocation(n_components=10)
modelling=model.fit(cv_model)
```

```python
#We use the get_feature_names() method from the CountVectorizerobject to get the list of feature names (i.e., words), and then
#use theargsort() method to sort the topics by the frequency of the words.Finally, we print the top 10 words in each topic.


feature_names = cv.get_feature_names()
for topic_idx, topic in enumerate(model.components_):
    print(f"Topic {topic_idx}:")
    print(" ".join([feature_names[i] for i in topic.argsort()[:-10 -1 :-1]]))
```

```python
#Evaluating....

print('Perplexity :',model.perplexity(cv_model))
```

Every Good Topic modelling.
 the result is 11.6 which Indicate that  The model will predicts so well to the unseen data, and a lower score also indicate better model.

```python
#having Done topic modelling On this data we can Carry out 

from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
```

```python
#initializing

sai=SentimentIntensityAnalyzer()
```

```python
#fittting in 


data={}

for i,rows in tqdm(eng_df.iterrows(),total=len(eng_df)):
    text=rows['original_title']
    row=rows['id']
    data[row]=sai.polarity_scores(text)
```

```python
#using vaders function to ceate the dataFrame and merging it with the actual dataframe

vaders=pd.DataFrame(data).T
vaders=vaders.reset_index().rename(columns={'index':'id'})
vaders=vaders.merge(eng_df,on='id',how='left')
```
